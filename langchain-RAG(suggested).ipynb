```python
from langchain.chat_models import ChatOpenAI
from langchain.document_transformers.openai_functions import create_metadata_tagger
from langchain.schema import Document
```

```python
llm = ChatOpenAI(temperature=0, model="gpt-3.5-turbo-0613", openai_api_key=openai_api_key)
schema = {
    "type": "object",
    "properties": {
        "section": {"type": "string"},
        "datasheet_title": {"type": "string"},
        "specification": {"type": "string"},
        "product parameters": {},
        "Product status links": {"type": "string"},
    },
    "required": ["section", "datasheet_title", "specification", "product parameters", "Product status links"]
}

# Assuming create_metadata_tagger is a function that uses the schema and llm to create a metadata tagger
document_transformer = create_metadata_tagger(metadata_schema=schema, llm=llm)
```

```python
print(document_transformer)
```

```python
original_documents = [
    Document(
        page_content=text)
]

enhanced_documents = document_transformer.transform_documents(original_documents)
```

```python
print(enhanced_documents)
```

```python
import json

def format_metadata(d):
    return json.dumps(d.metadata)

# Call the function for each document and join the results
formatted_metadata = "\n\n---------------\n\n".join(format_metadata(d) for d in enhanced_documents)

# Now you can use the 'formatted_metadata' variable whenever needed
print(formatted_metadata)
print(len(formatted_metadata))
print(type(formatted_metadata))


```

ChatBot

from now on We are going to use the chat bot technique in Langchian.
Chatbots use many of the tools we've already looked at with the addition of an important topic: Memory. There are a ton of different types of memory, tinker to see which is best for you.

```python
from langchain.llms import OpenAI
from langchain import LLMChain
from langchain.prompts.prompt import PromptTemplate

# Chat specific components
from langchain.memory import ConversationBufferMemory
```

```python
template = """
"You are a nice AI chat bot to help Electrical Engineers in selecting the right component for their circuits and recommendation. 
Please answer to the Enigneer's query based on the input data with precise"
datasheet informations are as follows:
{formatted_metadata}

{chat_history}
Human: {human_input}
Chatbot:"""

prompt = PromptTemplate(
    input_variables=["chat_history", "human_input"], 
    partial_variables={"formatted_metadata": formatted_metadata},
    template=template
)
memory = ConversationBufferMemory(memory_key="chat_history")
```

```python
llm_chain = LLMChain(
    llm=OpenAI(openai_api_key=openai_api_key),
    prompt = prompt,
    verbose=True,
    memory=memory
)
```

```python
llm_chain.predict(human_input="what is the maximum current allowed in this device?")
```

```python
llm_chain.predict(human_input="Can you explain me the general concept of this device?")
```
